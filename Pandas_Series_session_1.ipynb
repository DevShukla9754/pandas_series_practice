# Series from lists

```python
country = ['India','Pakistan','Nepal','Srilanka']

pd.Series(country)
```

```python
# Integers
runs = [13,24,45,67,100]

runs_series=pd.Series(runs)
runs_series
```

```python
# Custom index

marks=['67','57','89','100']
subjects = ['maths','english','science','hindi']

print(pd.Series(marks))
print('-----------------')
print(pd.Series(subjects))
print('-----------------')
print(pd.Series(marks,index=subjects))
print('-----------------')
print(pd.Series(subjects,index=marks))
print('-------------')
print('-------------')
marks_series_2 = pd.Series(subjects,index=marks)
```

```python
# Setting a name
marks=['67','57','89','100']
subjects = ['maths','english','science','hindi']

marks=pd.Series(marks,index=subjects,name='Dev ke marks')
marks
```

# Series from Dict :- keys index ban jata hai aur value,values ban jate hai

```python
marks = {
    'maths':67,
    'english':57,
    'science':89,
    'hindi':100
}
marks_series = pd.Series(marks,name = 'Dev ke marks')
marks_series
```

# Boolean Indexing on Series

```python
# Find the number of 50's and 100's score by kohli
print('-------------------')
print('-------------------')
print(kohli >= 50)           # ye ek boolean series return krega,mtlb jaha condition meet krega waha True otherwise False
print('-------------------')
print('-------------------')
print(kohli[kohli >= 50])     # yha upar wali boolean series ko as a index use kr rhe hai....jaha jaha true hoga wahi values return hogi bas
print('-------------------')
print('-------------------')
print('-------------------')
print(kohli[kohli >= 50].size)
```

```python
# Find the number of ducks

print(kohli[kohli == 0])
print('-------------------')
print('-------------------')
print(kohli[kohli ==0].size)
```

```python
# Count the number of day when I hasd more than 200 subs a day

subs[subs >= 200]
```

```python
subs[subs >= 200].size
```

```python
# Find the actors who have done more than 20 movies
num_counts = bolly.value_counts()    # value_counts() se return mein ek series hi milta hai
num_counts[num_counts >= 20]
```
